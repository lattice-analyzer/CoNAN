```python
import compiler
import deeo_decompser as deep

from compiler import AlgebraCompiler

############################################################################
# Parameters
############################################################################

n = 16
q = 1024
base_var = 2/3.

############################################################################
# Construct symbolic matrix
############################################################################

compilerobj = AlgebraCompiler(
    filename="examples.py",
    funcname='mul_NTRU',
    dimension=n
)


H = compilerobj.return_symbolic_matrix('LL')

############################################################################
# Construct lattice basis
############################################################################

L = compilerobj.return_lattice()



############################################################################
# Compute commutant basis
############################################################################

C = compilerobj.return_lattice('RL')


decomposer = deep.PrimaryDecomposer(
    lattice_basis=L,
    symbolic_matrix=H,
    commutant_basis=C,
    q=q,
    base_var=base_var,
    lattice_type="NTRU",

    ########################################################################
    # Factorization field
    #
    # "Q"    -> rational CRT decomposition
    # "Q[i]" -> Gaussian decomposition
    ########################################################################

    factor_field="Q",

    ########################################################################
    # Maximum regrouped blocks
    ########################################################################

    max_blocks=6
)

############################################################################
# Run search
############################################################################

result = decomposer.search(
    verbose=True
)

############################################################################
# Best attack
############################################################################

print("\nBest beta:")
print(result["best_beta"])

print("\nBest decomposition:")
print(result["best_result"])
```
