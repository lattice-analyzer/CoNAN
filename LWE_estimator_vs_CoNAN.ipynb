## The security analysis is performed under the classical sieving model, assuming a lattice reduction cost of $(2^{0.292\beta})$.

```python
################## NTRU composite #######################
n = 512
q = 7681
base_var = 4/3.

Xs = ND.DiscreteGaussian(math.sqrt(base_var))
Xe = ND.DiscreteGaussian(math.sqrt(base_var))
l = NTRU.Parameters(
    n=n,
    q=q,
    Xs=Xs,
    Xe=Xe
)


print("secruity according to the lattice Estimator: ",NTRU.primal_usvp(l)['beta']*0.292)


############################################################################
# Initialize estimator
############################################################################

estimator = SecurityEstimator(
    n=n,
    q=q,
    base_var=base_var,
    mul_function="mul_NTRU",
    level=1,    #### fast decomposer
    lattice="NTRU",
    mul_type="LL",
    filename="examples.py"
)

print("Checking dimension reduction homomorphisms, may take few minutes!")
############################################################################
# Run estimation
############################################################################

result = estimator.estimate_security(
    verbose=True
)

############################################################################
# Print result
############################################################################

print(result)
print("security according to CoNAN: ", result['concrete_beta']*0.292)
```

```python
################## Indocrypt ##################
# The dimension-reduction attack was already discussed in the original paper.
# However, the proposed approach requires solving three lattice instances,
# whereas our method achieves the same objective by solving only two lattices.
# This yields a practical improvement, although not an asymptotic one.
################################################
N = 139
q = 2237
base_var = 1.1

n = 6*N
Xs = ND.DiscreteGaussian(math.sqrt(1.1))
Xe = ND.DiscreteGaussian(math.sqrt(1.1))

l = NTRU.Parameters(
    n=n,
    q=q,
    Xs=Xs,
    Xe=Xe
)



print("security according to the lattice Estimator: ", NTRU.primal_usvp(l)['beta']*0.292)
############################################################################
# Initialize estimator
############################################################################

estimator = SecurityEstimator(
    n=n,
    q=q,
    base_var=base_var,
    mul_function="mul_NTRU3_ETRU",
    level=1,    #### fast decomposer
    lattice="NTRU",
    mul_type="LL",
    filename="examples.py"
)

print("Checking dimension reduction homomorphisms, may take few minutes!")
############################################################################
# Run estimation
############################################################################

result = estimator.estimate_security(
    verbose=True
)

############################################################################
# Print result
############################################################################
print(result)
print("security according to CoNAN: ", result['concrete_beta']*0.292)
```
