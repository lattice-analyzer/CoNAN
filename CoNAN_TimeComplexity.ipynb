### Algorithm: Detection of Dimension-Reducing Homomorphisms
---
##### Input
- Symbolic matrix $M(H_0,\ldots,H_{k-1})$
- Maximum folding factor $k$
- Current block count $b$ (default $b=1$)
##### Output
- Set of change-of-basis matrices $\mathcal{T}$
---
##### Express the Basis
1. If $b \ge k$, return $\emptyset$.
2. Write
   $$
   \mathbf{M}=\sum_{i=0}^{k-1}H_i\mathbf{A}_i
   $$
3. Compute the commutant
   $$
   \mathcal{C}\leftarrow
   \left\{
   \mathbf{X}
   \mid
   \mathbf{X}\mathbf{A}_i
   =
   \mathbf{A}_i\mathbf{X},
   \ \forall i
   \right\}
   $$
4. Compute a basis
   $$
   \{\mathbf{C}_1,\ldots,\mathbf{C}_m\}
   $$
   of $\mathcal{C}$.
---
##### Idempotent Search Over Basis
1. For each basis element
   $$
   \mathbf{C}_j \in
   \{\mathbf{C}_1,\ldots,\mathbf{C}_m\}
   $$
   do:
2. If $\mathbf{C}_j$ is a scalar matrix, continue.
3. Compute the characteristic polynomial
   $$
   p(x)=\det(x\mathbf{I}-\mathbf{C}_j)
   $$
4. If $p(x)$ is a power of an irreducible polynomial, continue.
5. Factor
   $$
   p(x)=p_1(x)p_2(x)
   $$
   such that
   $$
   \gcd(p_1,p_2)=1
   $$
6. Compute polynomials $s(x),t(x)\in\mathbb{F}[x]$ satisfying
   $$
   s(x)p_1(x)+t(x)p_2(x)=1
   $$
7. Construct the idempotent projector
   $$
   \mathbf{P}=t(\mathbf{C}_j)\cdot p_2(\mathbf{C}_j)
   $$
8. Break the loop.
9. If no projector $\mathbf{P}$ was constructed, return $\emptyset$.
---
##### Recursive Decomposition
1. Compute a basis of
   $$
   \operatorname{Im}(\mathbf{P})
   $$
2. Extend it to a basis of the full space and construct the change-of-basis matrix $\mathbf{T}$.
3. Compute the block decomposition
   $$
   \mathbf{T}^{-1}\mathbf{M}\mathbf{T}
   =
   \begin{pmatrix}
   \mathbf{M}_0 & \mathbf{0} \\
   \mathbf{0} & \mathbf{M}_1
   \end{pmatrix}
   $$
4. Update
   $$
   b \leftarrow b+1
   $$
5. Compute
   $$
   \mathcal{T}_0
   \leftarrow
   \text{Algorithm applied to }
   (\mathbf{M}_0,k,b)
   $$
6. Update
   $$
   b \leftarrow b+|\mathcal{T}_0|
   $$
7. Compute
   $$
   \mathcal{T}_1
   \leftarrow
   \text{Algorithm applied to }
   (\mathbf{M}_1,k,b)
   $$
8. Return
   $$
   \{\mathbf{T}\}
   \cup
   \mathcal{T}_0
   \cup
   \mathcal{T}_1
   $$
---

### Time Complexity Analysis
Now we calculate the time complexity of our algorithm. <br> <br>


##### 1. Express the Basis -
To get the blocks, we divide the $n \times n$ matrix $\mathbf{M}$ into small blocks of smaller blocks of size $\frac{n}{s} \times \frac{n}{s}$. Since $s$ is very small, we consider $k$ and $s$ to be equivalent. Now, we traverse the blocks in $\mathbf{M}$ one by one and do a very simple check that the new block is a multiple of an old block or it's transpose or not. If we detect a total of $k$-distinct blocks this way, then our total complexity will be $O(k^2 \times \frac{n^2}{s^2}) \approx O(n^2)$. Thus, we get the blocks $H_i$ and we write $\mathbf{M}=\sum_iH_i\mathbf{A}_i$ where we treat $H_i$ as symbols and $\mathbf{A}_i$ are $s \times s$ matrices. (technically this is the Hadamard product)<br> <br>

However, if $s$ is large, then $k \approx s^2$ and $s \approx O(n)$, so the overall complexity will become $O(n^4)$. In such situation, it is better to consider the $\mathbf{M} = \tau(h) = \sum_i h_i\tau(b_i) \Rightarrow \mathbf{A}_i = \tau(b_i)$ as given in our paper which will need only $O(n^3)$ time to compute. From here on, we assume the latter representation for $\mathbf{M}$ and proceed with our analysis. Also, by Proposition-1, we can show that computation of the commutant can be done in $O(n^3)$ time. <br> <br>


##### 2. Idempotent search over the Basis and Recursive decomposition -
We analyse this problem in a slightly different approach. For every basis, since we are computing coprime polynomials and then again decomposing them recursively, we can alternatively say that we can factor the characteristic polynomial down to powers of irreducible factors (just like prime power factorization of integers), and then partition those factors in every possible way. Every such partition will give us a number of distinct idempotents whose kernels will correspond to each subspace in the decomposition of our $\mathbf{M}$. <br> <br>

Thus, for each $\mathbf{C}_i$ we do the following - <br>
a. Compute the characteristic polynomial - The fastest algorithms to do this include Hessenberg reduction, QR algorithm which have a complexity of $O(n^3)$ but these algorithms are numerically unstable, which is not suitable for our application. Thus we consider algorithms like Faddeev–LeVerrier and Samuelson–Berkowitz which take $O(n^4)$ time. Thus, we compute $p(x) = \mathrm{det}(x\mathbf{I}_n - \mathbf{C}_i)$. <br>
b. Factorize the characteristic polynomials into irreducibles - This is step is relatively cheap as the modern algorithms only take $O(n^2)$ time upto logarithmic factors. Thus, we compute $p(x) = p_1(x)^{e_1}.p_2(x)^{e_2}\cdots p_s(x)^{e_s}$. <br>
c. Now we compute $\mathbf{C}_{ij} = f_j(\mathbf{C}_i)^{e_j}$ for all $j$. By Proposition-2, a single one of them can be computed in $O(n^2d_je_j)$, thus the total computation can be done in $O(n^2\sum d_je_j) = O(n^3)$ time. <br>
d. Now we need to compute the basis of kernel of all $\mathbf{C}_{ij}$. Kernel computation takes $O(mnr)$ time by guassian elimination, where $r$ is the rank. In our case, by linear algebra it can be shown that rank of $\mathbf{C}_{ij}$ is $d_je_j$, a total time of $O(n^2\sum d_je_j) = O(n^3)$ is needed. <br>
e. Thus we have $\mathbf{M} = \mathbf{T}\mathbf{M}'\mathbf{T}^{-1}$, where $\mathbf{M}'$ has $s$-blocks, each corresponding to a distinct $p_j$ that was found by factorization. Suppose we want to have a amximum of $k$-partitions, then it can be done in $S(s,k)$, where $S$ is Sterling numbers of second kind and they are asymptotically $O(\frac{k^s}{k!})$. It is approximately $O(k^s)$ in our case, as $k$ is very small. Every such case will require guassian elimination to compute new basis resulting in a $O(n^3)$ time per iteration. Thus, total time need is $O(n^3 k^s)$. Now, for the field of rational numbers, for a random polynomial $p$, the number of irreducible factors $s \approx 1$ as $\mathbb{Q}$ is a infinite field. Also we consider $k \le 6$, we can safely assume $k^s = O(\sqrt{n})$. <br> <br>

Thus, the total time complexity is $T(n) = O(n^3) + n\times O(n^4 + n^2\mathrm{log}^c(n) + n^3+n^3 + n^3\sqrt{n}) = O(n^5)$.
